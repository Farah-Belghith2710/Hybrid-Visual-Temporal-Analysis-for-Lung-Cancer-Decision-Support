```python
import pandas as pd
import numpy as np
import os
```

```python
canc=pd.read_csv(r"D:\PFA 2eme\nlst_780\nlst780.idc.delivery.052821\nlst_780_canc_idc_20210527.csv")
ctab=pd.read_csv(r"D:\PFA 2eme\nlst_780\nlst780.idc.delivery.052821\nlst_780_ctab_idc_20210527.csv")
ctabc=pd.read_csv(r"D:\PFA 2eme\nlst_780\nlst780.idc.delivery.052821\nlst_780_ctabc_idc_20210527.csv")
prsn=pd.read_csv(r"D:\PFA 2eme\nlst_780\nlst780.idc.delivery.052821\nlst_780_prsn_idc_20210527.csv")
screen=pd.read_csv(r"D:\PFA 2eme\nlst_780\nlst780.idc.delivery.052821\nlst_780_screen_idc_20210527.csv")

print("canc shape:", canc.shape)
print("ctab shape:", ctab.shape)
print("ctabc shape:", ctabc.shape)
print("prsn shape:", prsn.shape)
print("screen shape:", screen.shape)
 
```

```python
# Tes 395 patients téléchargés
import os
dataset_path=r"D:\PFA 2eme\manifest-1773666886424\NLST"
patient_ids = [int(p) for p in os.listdir(dataset_path) 
               if os.path.isdir(os.path.join(dataset_path, p))]

print(f"Patients disponibles: {len(patient_ids)}")

# Filtrer chaque CSV
canc_f   = canc[canc['pid'].isin(patient_ids)] 
ctab_f   = ctab[ctab['pid'].isin(patient_ids)]
ctabc_f  = ctabc[ctabc['pid'].isin(patient_ids)]
prsn_f   = prsn[prsn['pid'].isin(patient_ids)]
screen_f = screen[screen['pid'].isin(patient_ids)]

print(f"canc filtré: {canc_f.shape}")
print(f"ctab filtré: {ctab_f.shape}")
print(f"ctabc filtré: {ctabc_f.shape}")
print(f"prsn filtré: {prsn_f.shape}")
print(f"screen filtré: {screen_f.shape}")

patient_ids=[int(p) for p in os.listdir(dataset_path) 
               if os.path.isdir(os.path.join(dataset_path, p))]

print(f"Total patients: {len(patient_ids)}")
print(patient_ids)
```

```python
# Colonnes sélectionnées
screen_cols = ['pid', 'study_yr', 'ctdxqual', 'techpara_kvp', 'techpara_fov']

ctab_cols = ['pid', 'study_yr', 'sct_long_dia', 'sct_perp_dia', 
             'sct_margins', 'sct_pre_att', 'sct_ab_desc']

ctabc_cols = ['pid', 'study_yr', 'sct_ab_gwth', 'sct_ab_attn', 
              'sct_ab_invg', 'visible_days']

prsn_cols = ['pid', 'age', 'gender', 'cigsmok', 'race',
             'de_stag', 'can_scr', 'candx_days', 'canc_free_days',
             'scr_res0', 'scr_res1', 'scr_res2']

# Préparer chaque fichier
screen_clean = screen_f[screen_cols].groupby(['pid','study_yr']).first().reset_index()
ctab_clean   = ctab_f[ctab_cols].groupby(['pid','study_yr']).agg('max').reset_index()
ctabc_clean  = ctabc_f[ctabc_cols].groupby(['pid','study_yr']).agg('max').reset_index()
prsn_clean   = prsn_f[prsn_cols]

# Fusionner
temporal = pd.merge(screen_clean, ctab_clean, on=['pid','study_yr'], how='left')
temporal = pd.merge(temporal, ctabc_clean, on=['pid','study_yr'], how='left')
temporal = pd.merge(temporal, prsn_clean, on='pid', how='left')

print("Shape finale:", temporal.shape)
print(temporal.head(6))
```

```python
years_per_patient = temporal.groupby('pid')['study_yr'].count()
print("Distribution années par patient:")
print(years_per_patient.value_counts().sort_index())

complete = years_per_patient[years_per_patient == 3].index.tolist()
print(f"\nPatients avec 3 années complètes: {len(complete)}")
```

```python
# Garder seulement les 395 patients avec 3 années complètes
years_per_patient = temporal.groupby('pid')['study_yr'].count()
complete = years_per_patient[years_per_patient == 3].index.tolist()
temporal_complete = temporal[temporal['pid'].isin(complete)].copy()
temporal_complete = temporal_complete.sort_values(['pid', 'study_yr']).reset_index(drop=True)
```

```python
print("\nValeurs manquantes par colonne:")
print(temporal_complete.isnull().sum())
```

```python
temporal_clean = temporal_complete.copy()

# Colonnes nodules → 0 si pas de nodule détecté
nodule_cols = ['sct_long_dia', 'sct_perp_dia', 'sct_margins', 
               'sct_pre_att', 'sct_ab_gwth', 'sct_ab_attn', 
               'sct_ab_invg', 'visible_days']
temporal_clean[nodule_cols] = temporal_clean[nodule_cols].fillna(0)

# de_stag → 0 si pas de cancer
temporal_clean['de_stag'] = temporal_clean['de_stag'].fillna(0)

# candx_days → valeur max si pas de cancer (survie complète)
max_days = temporal_clean['candx_days'].max()
temporal_clean['candx_days'] = temporal_clean['candx_days'].fillna(max_days)

# techpara_fov et sct_ab_desc → médiane
temporal_clean['techpara_fov'] = temporal_clean['techpara_fov'].fillna(
    temporal_clean['techpara_fov'].median())
temporal_clean['sct_ab_desc'] = temporal_clean['sct_ab_desc'].fillna(
    temporal_clean['sct_ab_desc'].median())

# Vérification
print("NaN restants:")
print(temporal_clean.isnull().sum())
print(f"\nShape: {temporal_clean.shape}")
```

```python
from sklearn.preprocessing import MinMaxScaler

cols_to_normalize = ['ctdxqual', 'techpara_kvp', 'techpara_fov',
                     'sct_long_dia', 'sct_perp_dia', 'sct_margins',
                     'sct_pre_att', 'sct_ab_desc', 'sct_ab_gwth',
                     'sct_ab_attn', 'sct_ab_invg', 'visible_days',
                     'age', 'candx_days', 'canc_free_days',
                     'scr_res0', 'scr_res1', 'scr_res2']

scaler = MinMaxScaler()
temporal_clean[cols_to_normalize] = scaler.fit_transform(
    temporal_clean[cols_to_normalize])

print("Normalisation OK")
print(f"Shape: {temporal_clean.shape}")
```

```python
# Labels
canc_df = canc[['pid', 'de_stag']].copy()
stage_map = {
    110.0: 'IA', 120.0: 'IB', 210.0: 'IIA', 220.0: 'IIB',
    310.0: 'IIIA', 320.0: 'IIIB', 400.0: 'IV'
}
canc_df['stage_label'] = canc_df['de_stag'].map(stage_map)

labels = temporal_clean.groupby('pid').first().reset_index()
labels['label_diagnostic'] = labels['can_scr'].apply(lambda x: 1 if x > 0 else 0)

stage_map_int = {
    0.0: 0, 110.0: 1, 120.0: 2, 210.0: 3,
    220.0: 4, 310.0: 5, 320.0: 6, 400.0: 7
}
labels['label_staging'] = labels['de_stag'].map(stage_map_int).fillna(0).astype(int)
labels['label_survie'] = labels['canc_free_days']

nodule_trend = temporal_clean.groupby('pid')['sct_long_dia'].agg(['first', 'last'])
nodule_trend['label_recovery'] = (nodule_trend['last'] > nodule_trend['first']).astype(int)
labels = pd.merge(labels, nodule_trend[['label_recovery']], on='pid', how='left')

print("Labels shape:", labels.shape)
print(f"Diagnostic → {labels['label_diagnostic'].value_counts().to_dict()}")
print(f"Staging    → {labels['label_staging'].value_counts().sort_index().to_dict()}")
print(f"Recovery   → {labels['label_recovery'].value_counts().to_dict()}")
```

```python
def augment_patient(patient_data, n_augments=5):
    augmented = []
    numeric_cols = ['ctdxqual', 'techpara_kvp', 'techpara_fov',
                    'sct_long_dia', 'sct_perp_dia', 'sct_margins',
                    'sct_pre_att', 'sct_ab_gwth', 'sct_ab_attn',
                    'sct_ab_invg', 'visible_days']
    for i in range(n_augments):
        aug = patient_data.copy()
        noise = np.random.normal(0, 0.02, aug[numeric_cols].shape)
        aug[numeric_cols] = (aug[numeric_cols] + noise).clip(0, 1)
        scale = np.random.uniform(0.95, 1.05)
        aug[numeric_cols] = (aug[numeric_cols] * scale).clip(0, 1)
        aug['pid'] = aug['pid'] + (i + 1) * 100000
        augmented.append(aug)
    return pd.concat(augmented)
```

```python
from sklearn.model_selection import train_test_split

# 1. Split sur les ORIGINAUX seulement
original_pids_list = temporal_clean['pid'].unique().tolist()
train_orig, temp_pids = train_test_split(original_pids_list, 
                                          test_size=0.2, random_state=42)
val_orig, test_orig = train_test_split(temp_pids, 
                                        test_size=0.5, random_state=42)

print(f"Train originaux: {len(train_orig)}")
print(f"Val originaux:   {len(val_orig)}")
print(f"Test originaux:  {len(test_orig)}")

# 2. Augmenter SEULEMENT le train
train_original_df = temporal_clean[temporal_clean['pid'].isin(train_orig)]
val_df = temporal_clean[temporal_clean['pid'].isin(val_orig)]
test_df = temporal_clean[temporal_clean['pid'].isin(test_orig)]

all_augmented = []
for pid in train_orig:
    patient_data = temporal_clean[temporal_clean['pid'] == pid]
    aug_data = augment_patient(patient_data, n_augments=5)
    all_augmented.append(aug_data)

train_df = pd.concat([train_original_df] + all_augmented, ignore_index=True)

print(f"\nTrain final: {train_df['pid'].nunique()} patients")
print(f"Val:         {val_df['pid'].nunique()} patients (originaux purs ✅)")
print(f"Test:        {test_df['pid'].nunique()} patients (originaux purs ✅)")
```

```python
import pickle
import os

save_path = r"D:\PFA 2eme\preprocessed_clinical"
os.makedirs(save_path, exist_ok=True)

train_df.to_csv(os.path.join(save_path, "train_clinical.csv"), index=False)
val_df.to_csv(os.path.join(save_path, "val_clinical.csv"), index=False)
test_df.to_csv(os.path.join(save_path, "test_clinical.csv"), index=False)
labels.to_csv(os.path.join(save_path, "labels.csv"), index=False)

with open(os.path.join(save_path, "scaler.pkl"), 'wb') as f:
    pickle.dump(scaler, f)

print("✅ Fichiers sauvegardés !")
print(f"  train_clinical.csv → {len(train_df)} lignes")
print(f"  val_clinical.csv   → {len(val_df)} lignes")
print(f"  test_clinical.csv  → {len(test_df)} lignes")
print(f"  labels.csv         → {len(labels)} patients")
```
